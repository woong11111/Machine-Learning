## 1. Loading Data Set for Boston House

```python
import pandas as pd
import numpy as np

from sklearn.metrics import r2_score

import matplotlib.pyplot as plt
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
from xgboost import plot_importance

from copy import copy
```

```python
from sklearn.datasets import load_boston

boston = load_boston()
data = boston['data']
target = boston['target']

data = pd.DataFrame(data)
target = pd.DataFrame(target)

data.columns = boston['feature_names']
target.columns = ['y']

df = pd.merge(data, target, left_index=True, right_index=True)
df.head()
```

## 3. Feature Engineering

#### a. 결측치 기준 수 컬럼 제거

```python
standard = 0.5

enough_columns = []

for column in df.columns:
    count = df[column].count()
    
    if count >= df.shape[0] * (1 - standard) :
        enough_columns.append(column)
        
print("유효 column :", len(enough_columns))
```

#### b. Oversampling

```python
df_486 = df[df.index < 486]
mean = df_486['y'].mean()
std = df_486['y'].std()
mean, std
```

```python
df_oversample = df_486[(df_486['y'] > mean + std) | (df_486['y'] < mean + std)]
```

```python
df_oversample.shape
```

```python
df_oversample = pd.concat([df_oversample] * 2, axis = 0)
df_total = pd.concat([df_oversample, df], axis=0)
df_total.reset_index(inplace=True, drop = True)
```

#### c. Catory, Numeric 구분

```python
df_total_1 = df_total[enough_columns]

#df_total_1_cat = df_total_1.select_dtypes(indclude=['object'])
#df_total_1_num = df_total_1.select_dtypes(exclude=['object'])
#df_total_1_gls = df_total_1_cat[df_total_1_cat.columns[:1]]
#df_total_1_cat = df_total_1_cat[df_total_1_cat.columns[1:]]
#df_total_1_cat_one_hot = pd.get_dummies(df_total_1_cat)
#df_total_1_one_hot = pd.concat([df_total_1_num, df_total_1_cat_one_hot], axis=1)
#df_total_1_one_hot.reset_index(inplace=True)

#y = df_total_1_one_hot['y']
#X = df_total_1_one.hot.drop(columns=['y', 'index'])
y = df_total_1['y']
X = df_total_1.drop(columns=['y'])
```

```python
X = X.values
y = y.values
```

```python
X.shape, y.shape
```

## 4. Modeling

```python
def train(X, y, learning_rate=0.05, early_stopping_rounds=100, 
          start_index=486, end_index=506, fold=5, validation_space=5):
    
    xgb1 = xgb.XGBRegressor(silent=False,
                             n_estimators=500,
                             learning_rate=learning_rate,
                             tree_method='gpu_exact',
                             max_depth=5,
                             min_child_weight=3,
                             gamma=0,
                             subsample=0.8,
                             colsample_bytree=0.8,
                             metrics='mae',
                             #nthread=1,
                             seed=2019
                            )
    
    X_train_fold = []
    y_train_fold = []
    
    X_test_fold = []
    y_test_fold = []
    
    prediction_train_fold = []
    prediction_test_fold = []
    
    r2_scores = []
    mae_scores = []
    mae_counts = []
    
    index_array = np.arange(len(X))
    total_length = end_index - start_index
    one_fold_length = total_length // fold
    
    for i in range(fold) :
        
        train_index = index_array[:start_index + one_fold_length*(i) - validation_space]
        validation_index = index_array[start_index + one_fold_length*i - validation_space:start_index+5*i]
        test_index = index_array[start_index + one_fold_length*(i):start_index+one_fold_length*(i+1)]
        
        X_train, X_val, X_test = X[train_index], X[validation_index], X[test_index]
        y_train, y_val, y_test = y[train_index], y[validation_index], y[test_index]
        
        print("FOLD", i+1)
        print("TRAIN SHAPE: ", X_train.shape,
              "VALIDATION SHAPE: ", X_val.shape,
              "TEST SHAPE: ", X_test.shape)
        
        
        xgb1.fit(X_train, y_train, early_stopping_rounds = early_stopping_rounds,
                 eval_set=[(X_val, y_val)],
                 eval_metric='mae')
        
        best_round = xgb1.get_booster().best_ntree_limit

        xgb2 = xgb.XGBRegressor(silent=False,
                                n_estimators=best_round,
                                learning_rate=learning_rate,
                                tree_method='gpu_exact',
                                max_depth=5,
                                min_child_weight=1,
                                gamma=0,
                                subsample=0.8,
                                colsample_bytree=0.8,
                                metrics='mae',
                                #nthread=1,
                                seed=2019
                               )
        
        X_val = np.repeat(X_val, 5, axis=0)
        y_val = np.repeat(y_val, 5, axis=0)
        
        X_train = np.concatenate([X_train, X_val], axis=0)
        y_train = np.concatenate([y_train, y_val], axis=0)
        
        xgb2.fit(X_train, y_train)
        
        X_train_fold.append(X_train)
        y_train_fold.append(y_train)
        prediction_train_fold.append(xgb2.predict(X_train))
        
        X_test_fold.append(X_test)
        y_test_fold.append(y_test)
        prediction = xgb2.predict(X_test)
        prediction_test_fold.append(prediction)
        
        r2_scores.append(r2_score(y_true=y_test, y_pred=prediction))
        mae_scores.append(np.mean(np.abs(y_test-prediction)))
        mae_counts.append(np.abs(y_test-prediction) > 5)
        
    return X_train_fold, y_train_fold, prediction_train_fold, X_test_fold, y_test_fold, prediction_test_fold, r2_scores, mae_scores, mae_counts 
```

```python
start_index = 486 + df_oversample.shape[0]
end_index = 506 + df_oversample.shape[0]
fold = 4
validation_space = 5
```

```python
X_train_fold, y_train_fold, prediction_train_fold, X_test_fold, y_test_fold, prediction_test_fold, r2_scores, mae_scores, mae_counts = train(X, y,
                               learning_rate=0.05,
                               early_stopping_rounds=50,
                               start_index=start_index,
                               end_index=end_index,
                               fold=fold,
                               validation_space=validation_space
                              )
```

## 5. Results

```python
real = []
prediction = []

for i in range(4) :
    real.extend(y_test_fold[i])
    prediction.extend(prediction_test_fold[i])
    
real = np.array(real)
prediction = np.array(prediction)

print("MAE: ", np.mean(np.abs(real-prediction)))
print("R2: ", r2_score(real, prediction))
```

```python
for i in range(4) :
    plt.figure(figsize=(15,5))
    plt.plot(y_test_fold[i], 'o', label='y')
    plt.plot(prediction_test_fold[i], 'o', label='prediction')
    plt.plot(y_test_fold[i] - prediction_test_fold[i], label = 'error')
    plt.legend()
    plt.ylim([-50, 50])
    plt.title(str(start_index + 5*i) + "_" + str(mae_scores[i]))
    plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(real, label='real', color = 'blue')
plt.plot(prediction, label='prediction', color = 'red')
plt.plot(real - prediction, label = 'error', color = 'green')
plt.ylim([-50, 50])
plt.legend()
plt.show()
```
