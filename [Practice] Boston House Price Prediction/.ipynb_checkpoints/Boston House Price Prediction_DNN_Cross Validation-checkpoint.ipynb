## 4. Modeling

```python
def create_model(input_dim, learning_rate=0.001):
    
    #initialize the constructor
    model = Sequential()
    model.add(Dense(128, input_dim=input_dim))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.5))
    
    for i in range(4):
        model.add(Dense(128, kernel_regularizer=regularizers.l2(0.5)))
        model.add(BatchNormalization())
        model.add(Activation('relu'))
        model.add(Dropout(0.2))
        
    model.add(Dense(1))
    #compile model
    rmsprop = optimizers.RMSprop(lr =learning_rate)
    model.compile(loss="mean_absolute_error", optimizer=rmsprop)
    
    return model
```

```python
i = 1
seed = 2019

models = []
historys_list = []

X_train_list = []
X_test_list = []
y_train_list = []
y_test_list = []

kfold = KFold(n_splits=5, shuffle = True, random_state=2019)

for train_index, test_index in kfold.split(X):
    print("Fold:", i)
    i+=1
    
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    scaler = StandardScaler().fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    
    model = create_model(X_train.shape[1])
    
    #model Training
    early_stop = EarlyStopping(monitor='val_loss', patience=20, mode='auto')
    callbacks_list = [early_stop]
    
    history=model.fit(X_train,
                      y_train,
                      epochs=300,
                      batch_size=32,
                      validation_data=(X_test, y_test),
                      callbacks=callbacks_list
                     )
    
    X_train_list.append(X_train)
    y_train_list.append(y_train)
    X_test_list.append(X_test)
    y_test_list.append(y_test)
    
    models.append(model)
    historys_list.append(history)
```

```python
train_loss = history.history['loss']
validation_loss = history.history['val_loss']

plt.plot(train_loss, label = 'train')
plt.plot(validation_loss, label = 'test')
plt.legend()
plt.show()
```

## 5. Results

```python
mae_scores = []
r2_scores = []
val_prediction_list = []
```

```python
for i in range(5) :
    print("Fold:", i+1)
    train_prediction = models[i].predict(X_train_list[i])
    val_prediction = models[i].predict(X_test_list[i])
    
    val_prediction_list.append(val_prediction)
    
    print("R2 score:", r2_score(y_true=y_train_list[i], y_pred=train_prediction.flatten()))
    print("R2 score:", r2_score(y_true=y_test_list[i], y_pred=val_prediction.flatten()))
    
    mae = np.mean(np.abs(val_prediction.flatten() - y_test_list[i]))
    count = np.sum(np.abs(val_prediction.flatten() - y_test_list[i]) > 0.1)
    prediction_direction = np.sign(val_prediction.flatten() - 1)
    real_direction = np.sign(y_test_list[i] - 1)
    
    r2_scores.append(r2_score(y_true=y_test_list[i], y_pred=val_prediction.flatten()))
    mae_scores.append(np.mean(np.abs(val_prediction.flatten() - y_test_list[i])))
    
    print("평균 MAE:", mae)
    print("MAE 차이 0.1 이상 개수:", count)
    print("MAE 차이 0.1 이상 비율:", count / len(y_test_list[i]) * 100)
    print("방향성:", np.mean(prediction_direction == real_direction) * 100)
    print("\n")
```

```python
np.mean(mae_scores), np.mean(r2_scores)
```

```python
for i in range(4) :
    plt.figure(figsize=(15,5))
    plt.plot(y_test_list[i], 'o', label='y')
    plt.plot(val_prediction_list[i].flatten(), 'o', label='prediction')
    plt.plot(y_test_list[i] - val_prediction_list[i].flatten(), label = 'error')
    plt.legend()
    plt.ylim([-20, 70])
    plt.title("Fold" + str(i+1) + "_" + str(mae_scores[i]))
    plt.show()
```
