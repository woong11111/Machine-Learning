### 4) Check Column Data Type & Missing Data

```python
def missing_data(data):
    total = data.isnull().sum()
    percent = (data.isnull().sum()/data.isnull().count()*100)
    tt = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    types = []
    for col in data.columns:
        dtype = str(data[col].dtype)
        types.append(dtype)
    tt['Types'] = types
    return(np.transpose(tt))
```

## 4. Feature Engineering

### 1) Extract Y/M from Time Series Columns

```python
df['&     (fecha_alta)_month'] = df['&     (fecha_alta)'].map(lambda x : 0.0 if x.__class__ is float else float(x.split('-')[1])).astype(np.int8)
df['&     (fecha_alta)_year'] = df['&     (fecha_alta)'].map(lambda x : 0.0 if x.__class__ is float else float(x.split('-')[0])).astype(np.int16)

df['1   (ult_fec_cli_1t)_month'] = df['1   (ult_fec_cli_1t)'].map(lambda x : 0.0 if x.__class__ is float else float(x.split('-')[1])).astype(np.int8)
df['1   (ult_fec_cli_1t)_year'] = df['1   (ult_fec_cli_1t)'].map(lambda x : 0.0 if x.__class__ is float else float(x.split('-')[0])).astype(np.int16)
```

```python
df.shape
```

### 2) Make Time Lag Data

```python
# 1)   list  
products = y_columns

# 2)     
def date_to_int(str_date):
    Y, M, D = [int(a) for a in str_date.strip().split("-")]
    int_date = (int(Y) - 2015) * 12 + int(M)
    return int_date

df['int_date'] = df['(fecha_dato)'].map(date_to_int).astype(np.int8)

# 3) int_date  1  lag ,  _prev  

df_lag = df.copy()
df_lag.columns = [col + '_prev' if col not in ['int_date', ' (ncodpers)'] else col for col in df.columns]
df_lag['int_date'] += 1

# 4)   lag  '(fecha_dato)', ' (ncodpers)'  . 
#    lag  int_date  1      .

df = df.merge(df_lag, on = ['int_date', ' (ncodpers)'], how = 'left')

# 5)      
del df_lag

# 6)          0 
for prod in products :
    prev = prod + '_prev'
    df[prev].fillna(0, inplace=True)

df.fillna(-99, inplace=True)
```

### 1) Divide into Train / Validation / Test Data SET

```python
tst_X = tst.drop(columns=y_columns)
tst_X.shape
```

```python
tst_X.head()
```

```python
X = trn_val_X.values
y = trn_val_y.values
X_test = tst_X.values

print(X.shape, y.shape, X_test.shape)
```

```python
tst_X.shape
```

### 2) Training 

```python
i = 1

prediction_test_fold = []

kfold = KFold(n_splits=5, shuffle=True, random_state=2019)


param = {'booster' : 'gbtree',
         'max_depth' : 3,
         'nthread' : 4,
         'num_class' : len(y_columns),
         'objective' : 'multi:softprob',
         'silent' : 1,
         'eval_metric' : 'mlogloss',
         'eta' : 0.1,
         'min_child_weight' : 10,
         'colsample_bytree' : 0.8,
         'colsample_bylevel' : 0.9,
         'seed' : 2019}



for train_index, val_index in kfold.split(X) :
    print("Fold :", i)
    i+=1
    
    X_train, X_val = X[train_index], X[val_index]
    y_train, y_val = y[train_index], y[val_index]
    

    print("Train Shape :", X_train.shape,
          "Validation Shape :", X_val.shape,
          "Test Shape :", X_test.shape)
    
    dtrn = xgb.DMatrix(X_train, label=y_train, feature_names = trn_val_X.columns)
    dval = xgb.DMatrix(X_val, label = y_val, feature_names = trn_val_X.columns)
    dtst = xgb.DMatrix(X_test, feature_names = trn_val_X.columns)
    
    xgb1 = xgb.train(param, dtrn, num_boost_round=100, evals = [(dtrn, 'train'), (dval, 'eval')], early_stopping_rounds = 20)
                 
    prediction = xgb1.predict(dtst)
    prediction_test_fold.append(prediction)
```
